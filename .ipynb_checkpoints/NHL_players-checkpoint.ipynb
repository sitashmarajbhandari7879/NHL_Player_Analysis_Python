```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows',650)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import seaborn as sns
from matplotlib import pyplot
from sklearn.preprocessing import StandardScaler

```

```python
#loading the csv files
df = pd.read_csv('NHL.csv',sep=',', parse_dates = ['Born','DftYr'],encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
# handling strutural erros with respect to the dataset
""""
converting string into numeric_categorical values Eg: for column Hand : Right is considered 0 and left is considered 1"""

df['Hand'] = pd.factorize(df.Hand)[0]
df['Position_numeric']= pd.factorize(df.Position)[0]
df['Country'] = pd.factorize(df.Cntry)[0]

```

```python
#dropping irrelevant and repetative column
df.drop(['Born','City','Pr/St','Cntry','Nat','DftYr','Last Name','First Name','Position','Team'],inplace = True, axis = 1)
df.head()
```

```python
#Checking for duplicates
df.duplicated(keep=False).sum()
```

```python
# handling missing data
'''number of missing value for each attributes is checked to decide how it can be handled'''
df.isnull().sum()
```

```python
#dropping column with more than 10 missing values
df = df.dropna(axis='columns',thresh=10)
df = df.dropna() #dropping all the NA values 
df.shape
```

```python
#degisnating the target varible
X = df.drop('Salary', axis = 1)
Y = df['Salary']
X.head()
```

```python
#separating datset into train and test

x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state = 0)
x_train.shape,x_test.shape
```

```python
#Visualtizing the dataset using heatmap to get generic picture of the correlation between the columns

plt.figure(figsize = (12,10))
cor = x_train.corr()
sns.heatmap(cor,annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
#feature selection
scaler = StandardScaler().fit(x_train)
X_std = scaler.transform(X)
model = LinearRegression().fit(x_train, y_train)#defining the model
importance = model.coef_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
#Selecting higly correlated features and removing the first feature that is correlated with other fearure
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j])> threshold:
                colname = corr_matrix.columns[i] #getting the name of column
                col_corr.add(colname)
    return col_corr

corr_features = correlation(x_train, 0.90)
len(set(corr_features))
```

```python
corr_features
```

```python
x_train.drop(corr_features, axis = 1)
x_test.drop(corr_features, axis = 1)
```

```python
# creating the linear regression model using the training set
reg = LinearRegression().fit(x_train, y_train)
reg.score(x_train, y_train)
```

```python
# predicting using testing set
y_pred = reg.predict(x_test)
```

```python
# the coefficients
print("Coefficients: \n", reg.coef_)
```

```python
#Mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```

```python
# printing coefficient of determination
print("Coefficient of determination: %.2f" % r2_score(y_test,y_pred))
```

## Bayesian Ridge Regression

```python
# creating the linear regression model using the training set
reg = BayesianRidge().fit(x_train, y_train)
reg.score(x_train, y_train)
y_pred = reg.predict(x_test)
```

```python
#Mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# printing coefficient of determination
print("Coefficient of determination: %.2f" % r2_score(y_test,y_pred))
```
