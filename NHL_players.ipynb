```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows',650)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)

```

```python
#loading the csv files
df = pd.read_csv('NHL.csv',sep=',', parse_dates = ['Born','DftYr'],encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
# handling strutural erros with respect to the dataset
# converting string into numeric_categorical values
df['Hand'] = pd.factorize(df.Hand)[0]
df['Position_numeric']= pd.factorize(df.Position)[0]
df['Country'] = pd.factorize(df.Cntry)[0]

#for comparing the columns
df2 = df[["Position",'Position_numeric']]
df2.head()
```

```python
df3 = df[["Cntry",'Country']]
df3.head()
```

```python
df.dtypes
```

```python
#dropping irrelevant and repetative column
df.drop(['Born','City','Pr/St','Cntry','Nat','DftYr','Last Name','First Name','Position','Team'],inplace = True, axis = 1)
df.head()
```

```python
#Checking for duplicates
df.duplicated(keep=False).sum()
```

```python
# handling missing data
# number of missing value for each attributes
df.isnull().sum()
```

```python
#dropping column with more than 10 missing values
df = df.dropna(axis='columns',thresh=10)
type(df)
```

```python
df.shape
```

```python
df = df.dropna()
df.shape
```

```python
#degisnating the target varible
X = df.drop('Salary', axis = 1)
Y = df['Salary']
X.head()
```

```python
#separating datset into train and test
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state = 0)
x_train.shape,x_test.shape
```

```python
import seaborn as sns
plt.figure(figsize = (12,10))
cor = x_train.corr()
sns.heatmap(cor,annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
#Selecting higly correlated features and removing the first feature that is correlated with other fearure
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j])> threshold:
                colname = corr_matrix.columns[i] #getting the name of column
                col_corr.add(colname)
    return col_corr

corr_features = correlation(x_train, 0.80)
len(set(corr_features))
```

```python
corr_features
```

```python
x_train.drop(corr_features, axis = 1)
x_test.drop(corr_features, axis = 1)
```

```python
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# creating the linear regression model using the training set
reg = LinearRegression().fit(x_train, y_train)
reg.score(x_train, y_train)
```

```python
# predicting using testing set
y_pred = reg.predict(x_test)
```

```python
# the coefficients
print("Coefficients: \n", reg.coef_)
```

```python
#Mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```

```python
# printing coefficient of determination
print("Coefficient of determination: %.2f" % r2_score(y_test,y_pred))
```
